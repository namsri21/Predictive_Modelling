```python
import pandas as pd
```

```python
d = {
    'HCP_ID':['a1','a1','a1','a1','a1','a1','a1',
             'a2','a2','a2','a2','a2','a2','a2',
             'a3','a3','a3','a3','a3','a3','a3'],
    'Send_date':['2022-06-02','2022-06-04','2022-06-07','2022-06-20','2022-06-22','2022-06-28','2022-07-01',
                '2022-06-02','2022-06-04','2022-06-07','2022-06-20','2022-06-22','2022-06-28','2022-07-01',
                '2022-06-02','2022-06-04','2022-06-07','2022-06-20','2022-06-22','2022-06-28','2022-07-01'],
    'Channel':['F2F','Viber','Online','F2F','Email','Email','F2F',
              'Viber','Viber','Online','F2F','Email','Offline','F2F',
               'F2F','Viber','Online','F2F','Email','Email','F2F']
}

# Creating a DataFrame
df = pd.DataFrame(d)
df
```

```python
list(df.HCP_ID.unique())
```

```python
df_final=pd.DataFrame()
for i in list(df.HCP_ID.unique()):
    df_int=pd.DataFrame()
    
    ##Taking communication details of 1 HCP at a time in df_int
    df_int=df[df.HCP_ID==i]
    df_int=df_int.reset_index()
    df_int.drop(['index'],axis=1,inplace=True)
    
    #Create a column to store the immediate next communication channel for that HCP
    df_int['immediate_next_channel']=df_int['Channel'].shift(-1)
    
    #Immediate bect channel for the last communication is set as final communication
    df_int.immediate_next_channel=df_int.immediate_next_channel.fillna('Final_Communication')
    
    #Creating a df_final where we will store details of all HCP
    df_final=pd.concat([df_final,df_int],axis=0)
    df_final=df_final.reset_index()
    df_final.drop(['index'],axis=1,inplace=True)
```
