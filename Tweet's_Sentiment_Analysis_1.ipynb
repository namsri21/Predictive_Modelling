```python
# Creating the corpus along with structuring data in a readable form
corpus = []
for review in tweets:
    review = review.lower()
    review = re.sub(r"that's","that is",review)
    review = re.sub(r"there's","there is",review)
    review = re.sub(r"what's","what is",review)
    review = re.sub(r"where's","where is",review)
    review = re.sub(r"it's","it is",review)
    review = re.sub(r"who's","who is",review)
    review = re.sub(r"i'm","i am",review)
    review = re.sub(r"she's","she is",review)
    review = re.sub(r"he's","he is",review)
    review = re.sub(r"u","you",review)
    review = re.sub(r"c","see",review)
    review = re.sub(r"hey","hi",review)
    review = re.sub(r"v","we",review)
    review = re.sub(r"k","ok",review)
    review = re.sub(r"da","the",review)
    review = re.sub(r"ipads","ipad",review)
    review = re.sub(r"iphones","iphone",review)
    review = re.sub(r"ny","any",review)
    review = re.sub(r"fb","facebook",review)
    review = re.sub(r"aabout","about",review)
    review = re.sub(r"iclouds","icloud",review)
    review = re.sub(r"ur","your",review)
    review = re.sub(r"r","are",review)
    review = re.sub(r"al","all",review)
    review = re.sub(r"na","no",review)
    review = re.sub(r"they're","they are",review)
    review = re.sub(r"who're","who are",review)
    review = re.sub(r"ain't","am not",review)
    review = re.sub(r"wouldn't","would not",review)
    review = re.sub(r"shouldn't","should not",review)
    review = re.sub(r"can't","can not",review)
    review = re.sub(r"dont","do not",review)
    review = re.sub(r"don't","do not",review)
    review = re.sub(r"couldn't","could not",review)
    review = re.sub(r"won't","will not",review)
    review = re.sub(r"gonna","going to",review)
    review = re.sub(r"http\S+", "", review)#removing http links
    review = re.sub(r"@[\w]*", "", review)#removing all the words starting with @ .....eg : @naman
    review = re.sub(r"\W"," ",review)#Removing Non-Word Characters : eg : "The film, '@Pulp Fiction' was ? released in % $ year 1994." ------>The film Pulp Fiction was released in year 1994
    review = re.sub(r"\d"," ",review)#removing digits
    review = re.sub(r"\s+[a-z]\s+"," ",review)#Removing a Single Character : eg : text = "The film Pulp Fiction     s was b released in year 1994" ---->  The film Pulp Fiction was released in year 1994
    review = re.sub(r"\s+[a-z]$"," ",review)#Removing Spaces from End of each word
    review = re.sub(r"^[a-z]\s+"," ",review)#Removing Spaces from Start of each word
    review = re.sub(r"\s+"," ",review)#Removing Multiple Spaces eg : "The film      Pulp Fiction      was released in   year 1994." ------>The film Pulp Fiction was released in year 1994.
    review = re.sub(r'^br$', ' ', review)
    review = re.sub(r'\W', ' ',review)#Removing Non-Word Characters
    review = re.sub(r'iphones', 'iphone',review)
    review = re.sub(r'transparant ', 'transparent ',review)
    review = re.sub(r'bout', 'about',review)
    review = review.rstrip() #to remove trailing whitespaces at the end of the string
    review = review.lstrip() #to remove leading whitespaces at the start of the string
    corpus.append(review) 
```

```python
import nltk
## Removing stopwords( here i am doing it before lemmetization as latter converts words like 'has' to 'ha' and so if stop words 
## removal done after lemmetization then 'ha' wont be captured)

from nltk.corpus import stopwords

for i in range(len(corpus)):
    words = nltk.word_tokenize(corpus[i])
    words = [word for word in words if word not in stopwords.words('english')]
    corpus[i] = ' '.join(words)
```

### Predicting probability for tweeking the cutoff and finding the ideal cut_off

```python
test_train_predicted_prob=pd.DataFrame()
```

```python
test_train_predicted_prob = pd.DataFrame(train_tree.predict_proba(test_X))[[1]]
test_train_predicted_prob.columns = ['prob']
actual=test_y.reset_index()
actual.drop('index',axis=1,inplace=True)
# making a DataFrame with actual and prob columns
test_train_predict = pd.concat([actual, test_train_predicted_prob], axis=1)
test_train_predict.columns = ['actual','prob']
test_train_predict.head()
```

```python
test_train_predict.shape
```

```python
test_train_predict.tail()
```

```python
#Intution behind ROC curve - confusion matrix for each different cut-offs shows trade off in sensivity and specificity
test_roc_like_df=pd.DataFrame()
test_temp=test_train_predict.copy()
for cut_off in np.linspace(0,1,50):
    test_temp['predicted']=test_temp.prob.apply(lambda x:0 if x < cut_off else 1)
    test_temp['tp']=test_temp.apply(lambda x: 1 if x['actual']==1 and x['predicted']==1 else 0,axis=1)
    test_temp['fp']=test_temp.apply(lambda x: 1 if x['actual']==0 and x['predicted']==1 else 0,axis=1)
    test_temp['tn']=test_temp.apply(lambda x: 1 if x['actual']==0 and x['predicted']==0 else 0,axis=1)
    test_temp['fn']=test_temp.apply(lambda x: 1 if x['actual']==1 and x['predicted']==0 else 0,axis=1)
    sensitivity=test_temp['tp'].sum()/(test_temp['tp'].sum()+test_temp['fn'].sum())
    specificity=test_temp['tn'].sum()/(test_temp['tn'].sum()+test_temp['fp'].sum())
    test_roc_like_table=pd.DataFrame([cut_off,sensitivity,specificity]).T
    test_roc_like_table.columns=['cut_off','sensitivity','specificity']
    test_roc_like_df=pd.concat([test_roc_like_df,test_roc_like_table],axis=0)  
test_temp.sum()
plt.subplots(figsize=(10,4))
plt.scatter(test_roc_like_df['cut_off'],test_roc_like_df['sensitivity'],marker='*',label='Sensitivity',)
plt.scatter(test_roc_like_df['cut_off'],test_roc_like_df['specificity'],marker='o',label='Specificity')
#plt.scatter(test_roc_like_df['cut_off'],1-test_roc_like_df['specificity'],marker='*',label='FPR')
plt.title('For each cutoff, pair of sensitivity and FPR is plotted for ROC')
plt.legend()
```
